```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = "WorldCupMatches.csv"
df = pd.read_csv(dataset)
df.head()
```

```python
sns.countplot( x = df['Home Team Goals'] )
```

```python
sns.distplot(df['Year'])
```

```python
sns.boxplot(df['Year'])
```

```python
sns.scatterplot(x= df['Year'], y = df['Away Team Goals']  , hue = df['Home Team Goals'] , style = df['Half-time Home Goals'] , size = df['Half-time Away Goals'])
plt.show ()
df['Year'].min()
```

```python
sns.barplot (x = df['Away Team Goals'] , y = df['Year'] , hue = df['Half-time Home Goals'])
```

```python
sns.boxplot( x = df['Away Team Goals'] , y = df['Half-time Home Goals'] , hue = df['Datetime'] )
```

```python
import warnings
warnings.filterwarnings('ignore')
df['Year'].mean()

sns.distplot( df[df['Away Team Goals'] ==  'Right' ]['Year'], hist = False) #pdf of that person  who is righty
sns.distplot( df[df['Away Team Goals'] ==  'Left' ]['Year'] , hist = False) #pdf of that person  who is lefty
```

```python
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
print(df['Year'].dtype)

skill_X_preferredfoot  = pd.crosstab( df['Half-time Home Goals'], df['Away Team Goals'])
sns.heatmap(skill_X_preferredfoot)
plt.show()

new = df.groupby('Away Team Goals')['Year'].mean()
new.plot(kind='bar')
plt.show()
```

```python
cluster = pd.crosstab(df['Away Team Goals'] ,df['Half-time Home Goals'] )
sns.clustermap(cluster)
```
