```python
#Impor library yang diperlukan
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import interact, IntSlider
```

```python
dataset = "kc_house_data.csv"
df = pd.read_csv(dataset)
df.head()
```

```python
df.info()
```

```python
unique_value_counts = df.nunique()
print(unique_value_counts)
```

```python
summary = df.describe()#Menghasilkan ringkasan statistik dari dataframe

summary#Menampilkan ringkasan statistik
```

```python
# Meloop melalui setiap kolom dalam dataframe
for column in df.columns:
    # Memeriksa apakah tipe data kolom adalah float64 (numerik)
    if df[column].dtype == 'float64':
        # Membuat gambar (figure) untuk plot
        plt.figure(figsize=(8, 6))

        # Membuat histogram dengan KDE (Kernel Density Estimation)
        sns.histplot(df[column], kde=True, color='skyblue')

        # Menambahkan judul plot yang mencakup nama kolom
        plt.title(f'Distribusi {column}', size=15)

        # Menambahkan label sumbu x dengan nama kolom
        plt.xlabel(column, size=12)

        # Menambahkan label sumbu y
        plt.ylabel('Frequency', size=12)

        # Menampilkan plot
        plt.show()
```

```python
# Membuat tata letak subplot dengan ukuran 3x3
fig, axes = plt.subplots(3, 3, figsize=(15, 15))

# Memilih kolom-kolom numerik dalam dataframe
numerical_columns = [column for column in df.columns if df[column].dtype == 'float64']

# Meloop melalui kolom-kolom numerik
for i, column in enumerate(numerical_columns):
    # Memeriksa apakah kita telah mencapai jumlah maksimum subplot (9 subplot dalam tata letak 3x3)
    if i < 9:
        # Membuat histogram dengan KDE (Kernel Density Estimation) menggunakan subplot yang sesuai
        sns.histplot(df[column], kde=True, color='khaki', ax=axes[i//3, i%3])

        # Menambahkan judul plot yang mencakup nama kolom
        axes[i//3, i%3].set_title(f'Distribusi {column}', size=12)

        # Menambahkan label sumbu x dengan nama kolom
        axes[i//3, i%3].set_xlabel(column, size=10)

        # Menambahkan label sumbu y
        axes[i//3, i%3].set_ylabel('Frequency', size=10)
    else:
        break  # Berhenti jika telah mencapai jumlah maksimum subplot

# Mengatur tata letak subplot
plt.tight_layout()

# Menampilkan plot
plt.show()
```

```python
# Membuat tata letak subplot dengan ukuran 3x3
fig, axes = plt.subplots(3, 3, figsize=(15, 15))

# Memilih kolom-kolom numerik dalam dataframe
numerical_columns = [column for column in df.columns if df[column].dtype == 'float64']

# Meloop melalui kolom-kolom numerik
for i, column in enumerate(numerical_columns):
    # Memeriksa apakah kita telah mencapai jumlah maksimum subplot (9 subplot dalam tata letak 3x3)
    if i < 9:
        # Membuat histogram dengan KDE (Kernel Density Estimation) menggunakan subplot yang sesuai
        sns.histplot(df[column], kde=True, color='khaki', ax=axes[i//3, i%3])

        # Menambahkan judul plot yang mencakup nama kolom
        axes[i//3, i%3].set_title(f'Distribusi {column}', size=12)

        # Menambahkan label sumbu x dengan nama kolom
        axes[i//3, i%3].set_xlabel(column, size=10)

        # Menambahkan label sumbu y
        axes[i//3, i%3].set_ylabel('Frequency', size=10)
    else:
        break  # Berhenti jika telah mencapai jumlah maksimum subplot

# Mengatur tata letak subplot
plt.tight_layout()

# Menampilkan plot
plt.show()

```

```python
# Meloop melalui setiap kolom dalam dataframe
for column in df.columns:
    # Memeriksa apakah tipe data kolom adalah float64 (numerik)
    if df[column].dtype == 'float64':
        # Membuat gambar (figure) untuk plot
        plt.figure(figsize=(8, 6))

        # Membuat boxplot dengan orientasi sumbu y
        sns.boxplot(y=df[column], color='khaki')

        # Menambahkan judul plot yang mencakup nama kolom
        plt.title(f'Boxplot {column}', size=15)

        # Menambahkan label sumbu y dengan nama kolom
        plt.ylabel(column, size=12)

        # Menampilkan plot
        plt.show()
```

```python
# Membuat tata letak subplot 3x3 dengan ukuran 15x15
fig, axes = plt.subplots(3, 3, figsize=(15, 15))

# Memilih kolom-kolom numerik dalam dataframe
numerical_columns = [column for column in df.columns if df[column].dtype == 'float64']

# Meloop melalui kolom-kolom numerik
for i, column in enumerate(numerical_columns):
    # Memeriksa apakah kita telah mencapai jumlah subplot maksimum (9 subplot dalam tata letak 3x3)
    if i < 9:
        # Membuat boxplot dengan orientasi sumbu x dan menggunakan subplot yang sesuai
        sns.boxplot(x=df[column], color='purple', orient='h', ax=axes[i//3, i%3])

        # Menambahkan judul plot yang mencakup nama kolom
        axes[i//3, i%3].set_title(f'Boxplot {column}', size=12)

        # Menambahkan label sumbu x dengan nama kolom
        axes[i//3, i%3].set_xlabel(column, size=10)

        # Menambahkan label sumbu y
        axes[i//3, i%3].set_ylabel('Value', size=10)
    else:
        break  # Berhenti jika telah mencapai jumlah subplot maksimum

# Mengatur tata letak subplot
plt.tight_layout()

# Menampilkan plot
plt.show()
```

```python
# Menentukan daftar fitur yang akan digunakan dalam pairplot
fitur = ['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'sqft_above', 'sqft_basement', 'lat', 'long', 'sqft_living15', 'sqft_lot15']

# Mengatur gaya plot menggunakan Seaborn dengan "ticks" style
sns.set(style="ticks")

# Membuat pairplot dari dataframe menggunakan fitur yang telah ditentukan
sns.pairplot(df[fitur])

# Menampilkan plot
plt.show()
```

```python
# Membuat gambar (figure) dengan ukuran 10x8
plt.figure(figsize=(10, 8))

# Membuat heatmap korelasi dari dataframe dengan beberapa opsi:
# - annot=True: Menampilkan angka di dalam sel heatmap yang mewakili nilai korelasi.
# - cmap='coolwarm': Memilih skema warna untuk heatmap.
# - fmt='.2f': Format angka di dalam sel dengan dua angka desimal.
# - linewidths=0.5: Menambahkan garis tepi antara sel heatmap.
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Menambahkan judul heatmap
plt.title('Heatmap Korelasi Fitur', size=15)

# Menampilkan plot
plt.show()
```
