```python
# Mengelompokkan fitur ke dalam tiga kategori: numerik, kategorikal, dan target
numerical_features = ['carat', 'depth', 'table', 'x', 'y', 'z']
categorical_features = ['cut', 'color', 'clarity']
target = 'price'

# Meloop melalui setiap fitur kategorikal dan membuat pie chart
for feature in categorical_features:
    feature_counts = df[feature].value_counts()
    plt.figure(figsize=(8, 6))  # Mengatur ukuran grafik
    plt.pie(feature_counts, labels=feature_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set3'))
    plt.axis('equal')  # Membuat lingkaran menjadi lingkaran sejati
    plt.title(f'Pie Chart for {feature}')  # Menambahkan judul
    plt.show()
```

```python
# Mengelompokkan fitur ke dalam tiga kategori: numerik, kategorikal, dan target
numerical_features = ['carat', 'depth', 'table', 'x', 'y', 'z']
categorical_features = ['cut', 'color', 'clarity']
target = 'price'

# Membuat subplots dengan 2 baris dan 3 kolom
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Meloop melalui setiap fitur numerik dan membuat histogram pada setiap subplot
for i, feature in enumerate(numerical_features):
    row = i // 3
    col = i % 3
    sns.histplot(df[feature], kde=True, color='skyblue', ax=axes[row, col])
    axes[row, col].set_title(f'Distribution of {feature}')
    axes[row, col].set_xlabel(feature)
    axes[row, col].set_ylabel('Frequency')

# Mengatur tata letak subplot
plt.tight_layout()

# Menampilkan plot
plt.show()
```

```python
# Mengelompokkan fitur ke dalam tiga kategori: numerik, kategorikal, dan target
numerical_features = ['carat', 'depth', 'table', 'x', 'y', 'z']
categorical_features = ['cut', 'color', 'clarity']
target = 'price'

# Menentukan jumlah kolom dalam subplot sesuai dengan jumlah fitur numerik
num_cols = len(numerical_features)

# Membuat tata letak subplot yang dinamis
num_rows = (num_cols + 2) // 3  # Membagi jumlah kolom dengan 3 dan membulatkan ke atas
fig, axes = plt.subplots(num_rows, 3, figsize=(15, num_rows * 5))

# Meloop melalui setiap fitur numerik dan membuat box plot pada setiap subplot
for i, feature in enumerate(numerical_features):
    row = i // 3
    col = i % 3
    sns.boxplot(data=df, x=feature, color='purple', orient='h', ax=axes[row, col])
    axes[row, col].set_title(f'Boxplot {feature}', size=12)
    axes[row, col].set_xlabel(feature, size=10)

# Mengatur subplot yang tidak digunakan menjadi tidak terlihat
for i in range(num_cols, num_rows * 3):
    row = i // 3
    col = i % 3
    fig.delaxes(axes[row, col])

# Mengatur tata letak subplot
plt.tight_layout()
plt.show()
```
